# Assignment 1

```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

Nigeria House Dataset

```python
data = pd.read_csv("https://raw.githubusercontent.com/64060akothapa/64060_-akothapa/main/Assignment_1/nigeria_houses_data.csv")
```

# Descriptive statistics

Mean of all the prices of the houses in Nigeria

```python
print('Mean : '+str(data.price.mean()))
```

Median of all the prices of the houses in Nigeria

```python
print('Median : '+str(data.price.median()))
```

Mode of all the prices of the houses in Nigeria

```python
print('Mode : '+str(data.price.mode()))
```

# Transformation - transforming prices to logarithmic of 10

```python
data.price = data.price.apply(math.log10)
```

```python
data
```

# Scatter Plot

```python
plt.scatter(data.price, data.title)
plt.title("Price of House Vs Type of House")
plt.xlabel("Price of House")
plt.ylabel("Type of House")
plt.show()
```
